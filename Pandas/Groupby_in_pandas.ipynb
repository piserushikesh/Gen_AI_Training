Groupby

```python
import pandas as pd

var = pd.DataFrame({ "Name": ["a","b","c","d","a","b","a","b","a","c","c","d"],
                      "S_1":[12,13,14,12,13,14,15,23,25,16,10,17],
                     "S_2":[23,24,25,26,27,28,29,30,25,34,35,29]
                     })
var
```

```python
var_new = var.groupby("Name")

var_new
```

```python
for x,y in var_new:
  print(x)
  print(y)
  print()
```

```python
var_new.get_group("a")
```

```python
var_new.max()
```

```python
var_new.min()
```

```python
var_new.mean()
```
