 ***Insert***

```python
import pandas as pd

var = pd.DataFrame({"A":[1,2,3,4,5],"B":[9,8,7,6,5]})

var
```

```python
var.insert(1,"python",var["A"])

var
```

```python
var.insert(2,"python_1",[11,12,13,14,15])

var
```

```python
var["python_12"] = var["A"][:3]
var
```

***Delete***

```python
var1 = var.pop("python_1")
var1
```
